```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
data = pd.read_excel('Auto Sales data.xlsx')
data
```

```python
#sum of sales for each ordernumber
df2 = data.groupby('ORDERNUMBER')['SALES'].sum()
df2=df2.reset_index()
print(df2)
plt.figure(figsize=(10,8),dpi=80)
plt.scatter(df2['ORDERNUMBER'],df2['SALES'])
plt.xticks(rotation=80)
plt.xlabel('ORDERNUMBER')
plt.ylabel('SALES')
plt.grid()
plt.show()
```

```python
df3 = data.groupby(['COUNTRY','ORDERNUMBER'])['SALES'].sum()
df3 = df3.reset_index()
print(df3)
```

```python
#total sales for each country
df33 = df3.groupby('COUNTRY')['SALES'].sum()
df33 = df33.reset_index()
print(df33)
plt.figure(figsize=(10,8),dpi=80)
plt.scatter(df33['COUNTRY'],df33['SALES'])
plt.xticks(rotation=90)
plt.xlabel('COUNTRY')
plt.ylabel('Total SALES')
plt.grid()
plt.show()
```

```python
#total number of orders for each country
df4 = df3.groupby('COUNTRY')['ORDERNUMBER'].count()
df4 = df4.reset_index()
df4.sort_values(by='ORDERNUMBER' , ascending=False, inplace=True)
print(df4)
mylabel = df4['COUNTRY']
plt.figure(figsize=(10,8),dpi=100)
plt.pie(df4['ORDERNUMBER'],labels=mylabel)
plt.legend(title='Countries',bbox_to_anchor=(1.2, 1), loc='upper left')
```

```python
plt.figure(figsize=(10,8), dpi=70)
sns.distplot(df4['ORDERNUMBER'])
plt.show()
```

```python
#total number of orders for each country
plt.figure(figsize=(10,8), dpi=70)
plt.scatter(df4['COUNTRY'],df4['ORDERNUMBER'])
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
